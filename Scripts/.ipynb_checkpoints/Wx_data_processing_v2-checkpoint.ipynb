```python
def xls_cleaner(file_2_read, datetime_name = ["Date and Time", "Timestamp",  "TIMESTAMP"] ):

    data_xls = pd.read_excel(file_2_read,  header=None, index_col=None)  # read  xlsx to memory
    data_xls_trimmed = data_xls.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
    Col_true = data_xls_trimmed[data_xls_trimmed.isin(datetime_name)]
    Col_true = Col_true.dropna(how='all')
    finally_frame = Col_true.dropna(axis = 1, how='all')
    rows_to_skip = finally_frame.index[0]
    index_col_number = finally_frame.columns[0]

    data = pd.read_excel(os.path.join(datapath, files1[3]),  skiprows=rows_to_skip,  index_col=index_col_number)  # read  xlsx to memory

    return data

```
